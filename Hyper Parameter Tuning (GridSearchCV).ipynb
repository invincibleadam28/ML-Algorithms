```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns= iris.feature_names)
df.head()
```

```python
df['target']= iris.target
df.head()
```

```python
df['target'] = df.target.apply(lambda x: iris.target[x])
df.head()
```

```python
x = iris.data
y = iris.target 
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear',C=10,gamma='auto'),x,y,cv=5)    
```

```python
cross_val_score(SVC(kernel='rbf',C=10,gamma='auto'),x,y,cv=5)    
```

```python
cross_val_score(SVC(kernel='rbf',C=20,gamma='auto'),x,y,cv=5)    
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(SVC(gamma='auto'),{
    'C': [1,10,20,30,100],
    'kernel': ['rbf','linear']
}, cv=5, return_train_score= False)

clf.fit(x,y)
result_SVC= pd.DataFrame(clf.cv_results_)
```

```python
result_SVC
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV

rs = RandomizedSearchCV(SVC(gamma='auto'),{
    'C': [1,10,20],
    'kernel': ['rbf','linear']
    },
    cv=5,
    return_train_score = False,
    n_iter = 2
)
```

```python
rs.fit(x,y)
res = pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
res
```
